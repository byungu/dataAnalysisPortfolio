```python
import pandas as pd
```

```python
df = pd.read_csv("data/소년_천명당_소년범죄발생건수_시도__20220613164117.csv", index_col=0, encoding='cp949', engine='python')
df
```

```python
data_df  = df[['2018','2018.1','2018.2','2019','2019.1','2019.2','2020', '2020.1', '2020.2']]
data_df.head()
```

```python
data_df.columns = ['2018 소년 천명당 소년범죄발생건수', '2018 소년범죄발생건수', '2018 소년인구', '2019 소년 천명당 소년범죄발생건수', '2019 소년범죄발생건수', '2019 소년인구',
                 '2020 소년 천명당 소년범죄발생건수', '2020 소년범죄발생건수', '2020 소년인구']
data_df.head()
```

```python
data_df = data_df.drop(['행정구역별'], axis=0)
data_df.head()
```

```python
data_df.rename(columns = {'2018 소년 천명당 소년범죄발생건수':'2018 소년 천명당 소년범죄발생률',
                        '2019 소년 천명당 소년범죄발생건수':'2019 소년 천명당 소년범죄발생률',
                        '2020 소년 천명당 소년범죄발생건수':'2020 소년 천명당 소년범죄발생률'}, inplace=True)
```

```python
# 2018 세종시 데이터가 없어 19년도 데이터 삽입
data_df.loc['세종특별자치시',['2018 소년 천명당 소년범죄발생률']] = 8.0
data_df.loc['세종특별자치시',['2018 소년범죄발생건수']] = 322
data_df.loc['세종특별자치시',['2018 소년인구']] = 40037
data_df 
```

```python
data_df = data_df.astype('float') 
```

```python
# 18, 19, 20 년도 범죄 발생률 컬럼 추가
data_df['소년범죄 발생률 3개년 평균'] = (data_df['2018 소년 천명당 소년범죄발생률'] + data_df['2019 소년 천명당 소년범죄발생률'] + 
                                                                      data_df['2020 소년 천명당 소년범죄발생률']) / 3
data_df.head()
```

```python
data_df['전국비율'] = data_df['2020 소년범죄발생건수'] / 64573 * 100
data_df
```
