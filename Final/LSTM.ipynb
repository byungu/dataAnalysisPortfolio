```python
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
```

```python
data = crim.columns
data2 = crim.index
sido = []
X = []
```

```python
for i in data2:
    X.append(i)
```

```python
for d in data:
    sido.append(d)
```

```python
X = [X]
predict2 = []
```

```python
def make_sample(data, window):
    train = []
    target = []
    for i in range(len(data)-window): # 전체 데이터 갯수에서 윈도우 값만큼 데이터를 얻으니 그만큼 제외한 범위를 갖는다.
        train.append(data[i:i+window]) # 0 + 7 이면 0~ 7까지 1 + 7이면 1~8까지 데이터를 담는다
        target.append(data[i+window]) # 타겟은 윈도우값을 더한 다음 값
    return np.array(train), np.array(target) # 얻은 값을 반환받는다.
```

```python
for i in range(17):
    sidocrim = (crim[sido[i]]).to_numpy()
    sidocrim = pd.DataFrame(sidocrim)
    scaler = MinMaxScaler(feature_range = (0, 1))
    scaled = scaler.fit_transform(sidocrim)
    TEST_SIZE = 1
    train_data = scaled[:-TEST_SIZE]
    test_data = scaled[-3:]
    data = X[-TEST_SIZE:]

    X_train, Y_train = make_sample(train_data, 2)
    model = Sequential()
    model.add(LSTM(1024, input_shape=(X_train.shape[1], 1), activation='tanh', return_sequences=False))
    model.add(Dense(512, activation='tanh'))
    model.add(Dense(256, activation='tanh'))
    model.add(Dense(128, activation='tanh'))
    model.add(Dense(64, activation='tanh'))
    model.add(Dense(1, activation='tanh'))
    
    model.compile(optimizer='adam', loss='mse', metrics=['accuracy'])
    model.fit(X_train, Y_train, epochs=100, batch_size=20)
    
    print(test_data)
    X_test, Y_test = make_sample(test_data, 2)
    
    pred = model.predict(X_test)
    pred = scaler.inverse_transform(pred)
    predict2.append(pred)
```

```python
predict2 = np.array(predict2)
```

```python
predict2 = predict2.reshape(-1)
```

```python
predict2 = pd.DataFrame(predict2)
```

```python
predict2 = round(predict2, 5)
```
